```python
students = [] ##list of students
    
def display_menu():
    print("1. Add Student")
    print("2. Remove Student")
    print("3. View all Students")
    print("4. Update a Student grade")
    print("5. Sort students by name or grade")
    print("6. Exit")

def add_student(): # Adding students details
    total_students=int(input("Enter the total number of students="))
    for i in range(1, total_students+1):
        name = input(f"Enter the name of student{i}=")
        grade = float(input(f"Enter the student's grade '{name}'="))
        students.append({"name":name , "grade":grade}) ##data is stored in dictionary 
        print(f"Student '{name}' added with grade '{grade}'")

def remove_student(): # Removing students details
    name = input("Enter the student name=")
    for student in students:
        if student["name"] == name:
            students.remove(student)
            print(f"Student '{name}' is removed")
            return
    print(f"Student '{name}' not found")

def view_students(): # View all details of students
    if not students:
        print("No students available.")
    else:
        for student in students:
            print(f"Name: {student['name']}, Grade: {student['grade']}")

def update_student():  # updating all detail of student 
    name = (input("Enter the student name you want to update his grade:"))
    for student in students:
        if student["name"] == name:
            new_grade = float(input(f"Enter the new grade for {name}:"))
            student["grade"] = new_grade
            print(f"Grade of student '{name}' is update to '{new_grade}'")
            return ## Exit the function after updating the grade
            
    print(f"Student {name} is not found")
            
    
def sort_student():
    criteria = input("Sort by 'name' or 'grade'")
    if criteria == "name":
        students.sort(key=lambda student:student["name"])
    elif criteria == "grade":
        students.sort(key=lambda student:student["grade"])
    else:
        print("Invalid criteria, please chose 'name' or 'grade'")
        return
    print("Sorted successfully")
    view_students()

def main():
    print("--Welcome to student management system--") 
    while True:
        display_menu()
        choice = input("Enter the choice: ")
        if choice == "1":
            add_student()
        elif choice == "2":
            remove_student()
        elif choice == "3":
            view_students()
        elif choice == "4":
            update_student()
        elif choice == "5":
            sort_student()
        elif choice == "6":
            print("Exiting Student Management System.")
            break
        else:
            print("Invalid choice. Please try again")

#call the main function
main()
```
